```python
def coordlab():
    coor = {
        's': [1, 1],'a': [2, 3],'b': [2, 4],'c': [2, 2],'d': [1, 4],'j': [3, 4],'e': [3, 2],
        'f': [3, 1],'g': [4, 4],'i': [3, 3]
    }
    return coor
```

3. FUNCIÓN h(N, NOB) EN PYTHON, BASADA EN LA FÓRMULA

```python
def h(N, NOB):
    from math import fabs
    coord = coordlab()

    return fabs(coord[NOB][0] - coord[N][0]) + fabs(coord[NOB][1] - coord[N][1])
```

4. FUNCIÓN formarh(ABIERTOS , NOB), QUE FORME UNA LISTA CON h(Ni, NOB) DE CADA NODO Ni DE ABIERTOS.

```python
def formarh(abiertos, nob):
    lish = []
    for ni in abiertos:
        lish.append(h(ni, nob))
    return lish
```

```python
print(formarh(['s','a','b'], 'g'))
```

5. FUNCION menor(LISH), QUE RETORNA LA POSICIÓN QUE OCUPA EL MENOR DE LISH

```python
def menor(lish):
    me = lish[0]
    pos = 0

    for i in range(len(lish)):
        if lish[i] < me:
            me = lish[i]
            pos = i

    return pos
```

```python
print(menor([6, 2, 3]))
```

6. BUSQUEDA PRIMERO EL MEJOR PARA EL LABERINTO

```python
def bpm(nin, nob, graf):
    cerrados = []
    abiertos = [nin]

    while abiertos:
        n = abiertos[menor(formarh(abiertos, nob))]
        cerrados.append(n)

        if n == nob:
            return cerrados

        abiertos.remove(n)

        for v in graf[n]:
            if v not in cerrados and v not in abiertos:
                abiertos.append(v)

    return 'falla'
```

```python
print(bpm('s', 'g', laberinto()))
```

7. PROBLEMA. BÚSQUEDA DE LA ESCALADA EN PYTHON (APLICARLA AL PROBLEMA DEL LABERINTO)

```python
def escalada(nin, nob, graf):
    cerrados = []
    n = nin

    while n != nob:
        cerrados.append(n)

        vecinos = graf[n]

        # si no hay vecinos, falla
        if not vecinos:
            return "falla"

        # elegir mejor vecino (menor h)
        h_vals = formarh(vecinos, nob)
        mejor = vecinos[menor(h_vals)]

        # condición de escalada
        if h(mejor, nob) > h(n, nob):
            return "falla"

        n = mejor

    cerrados.append(n)
    return cerrados
```

```python
print("Escalada:")
print(escalada('s', 'g', laberinto()))
```
