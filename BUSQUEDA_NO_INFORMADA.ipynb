PROB1. REPRESENTAR EL SIGUIENTE ARBOL MEDIANTE UN DICCIONARIO EN PYTHON


image.png![image.png]()

```python
def arbol():
    ar = {
        'S': ['A', 'B'],
        'A': ['C', 'D'],
        'B': ['E', 'F'],
        'C': ['H', 'G'],
        'D': [],
        'E': [],
        'F': [],
        'H': [],
        'G': ['I', 'J'],
        'I': [],
        'J': []
    }
    return ar
```

PROB 2. APLICAR LA BÚSQUEDA HORIZONTAL EN PYTHON AL ÁRBOL DEL PROBLEMA 1.

```python
def bha(nin, nob, arb):
    cerrados = []
    abiertos = [nin]

    while abiertos:
        n = abiertos[0]
        cerrados.append(n)

        if n == nob:
            return cerrados

        abiertos.remove(n)
        hijos = arb[n]
        abiertos = abiertos + hijos

    return 'falla'

print(bha('S', 'J', arbol()))
```

PROB 3. APLICAR LA BÚSQUEDA VERTICAL EN PYTHON AL ÁRBOL DEL PROBLEMA 1

```python
def bva(nin, nob, arb):
    cerrados = []
    abiertos = [nin]

    while abiertos:
        n = abiertos[0]
        cerrados.append(n)

        if n == nob:
            return cerrados

        abiertos.remove(n)
        hijos = arb[n]
        abiertos = hijos + abiertos

    return 'falla'


print(bva('S', 'J', arbol()))
```

PROB 4. REPRESENTAR EL SIGUIENTE GRAFO MEDIANTE UN DICCIONARIO EN PYTHON
image.png


```python
grafo = {
    'S': ['A', 'B'],'A': ['S', 'C', 'D'],'B': ['S', 'D', 'E'],'C': ['A'],'D': ['A', 'B'],'E': ['B']
}
```

PROB 5. APLICAR LA BÚSQUEDA HORIZONTAL EN PYTHON AL GRAFO DEL PROBLEMA 4.

```python
def bhg(nin, nob, graf):
    cerrados = []
    abiertos = [nin]

    while abiertos:
        n = abiertos[0]
        cerrados.append(n)

        if n == nob:
            return cerrados

        abiertos.remove(n)

        for v in graf[n]:
            if v not in cerrados and v not in abiertos:
                abiertos.append(v)

    return 'falla'

print(bhg('S', 'E', grafo))
```

PROB 6. APLICAR LA BÚSQUEDA VERTICAL EN PYTHON AL GRAFO DEL PROBLEMA 4

```python
def bvg(nin, nob, graf):
    cerrados = []
    abiertos = [nin]

    while abiertos:
        n = abiertos.pop(0)
        cerrados.append(n)

        if n == nob:
            return cerrados

        for v in reversed(graf[n]):
            if v not in cerrados and v not in abiertos:
                abiertos.insert(0, v)

    return 'falla'

print(bvg('S', 'E', grafo))
```
