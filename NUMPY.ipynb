Mtodos array

```python
import numpy as np

arr = np.arange(10, 20)

div_por_3 = arr % 3 == 0

print(arr.sum())  # sumatoria
print(arr.mean())  # promedio
print(arr.std())  # desviación estándar
print(arr.max())  # mayor
print(arr.min())  # menor

print(div_por_3.all())  # todos son True
print(div_por_3.any())  # alguno es True
print(div_por_3.sum())  # sumatoria (No.) de True
print(div_por_3.nonzero())  # posiciones de True
```

Métodos array - ordenar

```python
import numpy as np

a = np.array([4.5, 2.3, 6.7, 1.2, 1.8, 5.5])

a.sort()    # ordena el array físicamente, lo modifica
print(a)

x = np.array([4.5, 2.3, 6.7, 1.2, 1.8, 5.5])

print(np.sort(x))   # ordena sin modificar el array x
print(x)

s = x.argsort()   # indice de elementos por orden
print(s)

print(x[s])   # muestra x ordenado, según s
```

3. CON NUMPY, DADAS DOS MATRICES m1 Y m2, CALCULAR Y MOSTRAR (VERIFICANDO QUE SE CUMPLAN LAS CONDICIONES):

 s = m1+m2

 p = m1*m2

 inversa de m1

 determinante de m2

```python
import numpy as np
```

```python
m1 = np.array([[1, 2], [3, 4]], dtype=float)
m2 = np.array([[5, 6], [7, 8]], dtype=float)

print("m1 =\n", m1)
print("m2 =\n", m2)
```

```python
# Para suma y producto elemento a elemento
if m1.shape == m2.shape:
    print("✔ Matrices con misma dimensión")
else:
    print("X Error: dimensiones diferentes")

# Para inversa y determinante
if m1.shape[0] == m1.shape[1]:
    print("✔ m1 es cuadrada")
else:
    print("X m1 no es cuadrada")

if m2.shape[0] == m2.shape[1]:
    print("✔ m2 es cuadrada")
else:
    print("X m2 no es cuadrada")
```

```python
s = m1 + m2
print("Suma:\n", s)
```

```python
p = m1 * m2
print("Producto elemento a elemento:\n", p)
```

```python
det_m1 = np.linalg.det(m1)

if det_m1 != 0:
    inv_m1 = np.linalg.inv(m1)
    print("Inversa de m1:\n", inv_m1)
else:
    print("X m1 no tiene inversa")
```

```python
det_m2 = np.linalg.det(m2)
print("Determinante de m2:", det_m2)
```
