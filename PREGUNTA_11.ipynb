```python
grafo = {
    'S': ['A', 'B'],
    'A': ['S', 'C', 'D'],
    'B': ['S', 'E', 'F'],
    'C': ['A', 'D'],
    'D': ['A', 'C'],
    'E': ['B', 'F'],
    'F': ['B', 'E']
}
```

```python
from collections import deque

def bfs(grafo, inicio, objetivo):
    visitados = set()
    cola = deque([[inicio]])  # guardamos caminos

    while cola:
        camino = cola.popleft()
        nodo = camino[-1]

        if nodo == objetivo:
            return camino

        if nodo not in visitados:
            visitados.add(nodo)

            for vecino in grafo[nodo]:
                nuevo_camino = list(camino)
                nuevo_camino.append(vecino)
                cola.append(nuevo_camino)

    return None
```

```python
resultado = bfs(grafo, 'S', 'E')
print("Camino encontrado:", resultado)
```
